```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

# y=w0 + xw1 + (x^2)w2 => y= w0 + x1w1 + x2w2 
X=np.array([[-3,-2,-1,0,1,2,3]]).T
y=np.array([[7.5,3,0.5,1,3,6,14]]).T
X2=X**2 

ones=np.ones((X.shape[0],1))
Xbar=np.concatenate((ones,X),axis=1)
Xbar=np.concatenate((Xbar,X2),axis=1)

A=np.dot(Xbar.T,Xbar)
b=np.dot(Xbar.T,y)
w=np.dot(np.linalg.pinv(A),b)
# w=[w_0;w_1;w_2]
w_0=w[0][0]
w_1=w[1][0]
w_2=w[2][0]
x0=np.linspace(-5,5,50)
y0=w_0 + w_1*x0 + w_2*x0**2
plt.plot(X,y,'ro')
plt.plot(x0,y0)
plt.axis([-4,4,0,20])
plt.xlabel('x')
plt.ylabel('y')
```
